# Imports

```python
import math
import pandas as pd
import nfl_data_py as nfl
```

# Creating Mass Data Set

```python
years = range(2000, 2023 + 1)
```

## Seasonal Data

```python
df_seasonal = nfl.import_seasonal_data(years)
```

```python
df_seasonal.head()
```

## Combine and Draft Capital

```python
df_combine = nfl.import_combine_data(years)
```

```python
df_combine.head()
```

```python
df_draft_cap = nfl.import_draft_picks(years)
```

```python
df_draft_cap.head()
```

## Roster Data

```python
df_roster = nfl.import_seasonal_rosters(years)
```

```python
df_roster.head()
```

### Massaging Roster Data

```python
df_roster_seasonal = df_roster[["position", "player_name", "player_id", "pfr_id", "season", "age"]]
```

## Merging DFs

```python
df_merge = df_seasonal.merge(df_roster_seasonal, on=["player_id", "season"], how="left")
```

```python
df_merge_cleaned = df_merge.copy().drop_duplicates()
```

```python
# for col in df_combine.columns:
#     print(f"df_merge_cleaned.at[i, '{col}'] = combine_row['{col}'].iloc[0]")
```

```python
for i, row in df_merge_cleaned.iterrows():
    pfr_id = row["pfr_id"]

    
    combine_row = df_combine.query(f"""pfr_id == "{pfr_id}" """)    
    if not combine_row.empty:
        # draft info
        df_merge_cleaned.at[i, 'draft_year'] = combine_row['draft_year'].iloc[0]
        df_merge_cleaned.at[i, 'draft_team'] = combine_row['draft_team'].iloc[0]
        df_merge_cleaned.at[i, 'draft_round'] = combine_row['draft_round'].iloc[0]
        df_merge_cleaned.at[i, 'draft_ovr'] = combine_row['draft_ovr'].iloc[0]
        df_merge_cleaned.at[i, 'school'] = combine_row['school'].iloc[0]
        
        # combine info
        df_merge_cleaned.at[i, 'ht'] = combine_row['ht'].iloc[0]
        df_merge_cleaned.at[i, 'wt'] = combine_row['wt'].iloc[0]
        df_merge_cleaned.at[i, 'forty'] = combine_row['forty'].iloc[0]
        df_merge_cleaned.at[i, 'bench'] = combine_row['bench'].iloc[0]
        df_merge_cleaned.at[i, 'vertical'] = combine_row['vertical'].iloc[0]
        df_merge_cleaned.at[i, 'broad_jump'] = combine_row['broad_jump'].iloc[0]
        df_merge_cleaned.at[i, 'cone'] = combine_row['cone'].iloc[0]
        df_merge_cleaned.at[i, 'shuttle'] = combine_row['shuttle'].iloc[0]
    else:
        draft_capital_row = df_draft_cap.query(f"""pfr_player_id == "{pfr_id}" """) 
        
        if not draft_capital_row.empty:
            df_merge_cleaned.at[i, 'draft_year'] = draft_capital_row['season'].iloc[0]
            df_merge_cleaned.at[i, 'draft_team'] = draft_capital_row['team'].iloc[0]
            df_merge_cleaned.at[i, 'draft_round'] = draft_capital_row['round'].iloc[0]
            df_merge_cleaned.at[i, 'draft_ovr'] = draft_capital_row['pick'].iloc[0]
```

# Adding Fantasy Tiering

```python
df_fantasy_tiering = df_merge_cleaned.copy()
```

```python
def assign_tier(group):
    group['tier'] = (group.reset_index(drop=True).index // 12) + 1
    return group
```

## WRs

```python
df_wr = df_fantasy_tiering.loc[df_fantasy_tiering["position"] == "WR"]
```

```python
df_sorted_wr = df_wr.copy().sort_values(by=['season', 'fantasy_points_ppr'], ascending=[True, False])
df_tiered_wr = df_sorted_wr.groupby('season').apply(assign_tier)
```

```python
df_tiered_wr.query("season == 2023")[["player_name", "fantasy_points_ppr",
                                      "tier", "draft_ovr", "draft_team"]].head(36)
```
