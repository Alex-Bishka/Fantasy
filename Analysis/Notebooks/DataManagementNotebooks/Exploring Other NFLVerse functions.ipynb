## Snap Percentage

```python
snap_years = range(2012, 2023 + 1)
df_snaps = nfl.import_snap_counts(snap_years)
```

```python
df_snaps.head()
```

```python
df_snaps.query("player == 'Mike Evans'") \
    .groupby(by='season') \
    .agg({"offense_snaps": "sum", "offense_pct": "mean"})
```
