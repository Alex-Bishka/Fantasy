# Merging the two

```python
# Step 1: Prepare the new DataFrame from the dictionary
data = []
for key, df in team_stats_dict.items():
    first_row = df.iloc[0]  # Assuming each DataFrame has at least one row
    data.append({
        'key': key,
        'Att_passing_totals': first_row['Att_passing_totals'],
        'Att_rushing_totals': first_row['Att_rushing_totals'],
        'total_snaps': first_row['total_snaps']
    })
```

```python
# Create a DataFrame from the list
stats_df = pd.DataFrame(data)
stats_df.head()
```

```python
# Step 2: Prepare the key in df_core
df_core['key'] = df_core['team'] + '-' + df_core['season'].astype(str)
```

```python
# Step 3: Merge the DataFrames
df_merge = df_core.merge(stats_df, on='key', how='left')
```

```python
df_merge.query("season >= 2012").head()
```

```python
# Optionally, you might want to drop the 'key' column if it's no longer needed
df_merge.drop('key', axis=1, inplace=True)
```

# Save Merged df

```python
df_merge.to_pickle("../../FantasyData/data-frames/df_basic_ngs_snaps_adv_redzone-tgts_team-snap-data_1999_2023.pkl")
```
